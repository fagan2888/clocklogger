```python
%matplotlib inline
import numpy as np
from scipy import signal
```

This is a sample of what we're aiming for:

```python
example_data = np.load('../../../Dropbox/code/dataq/record_20130330_2358_15s.npz')
yex = -example_data['signal']
sampling_rate = example_data['fs'][()]
tex = np.arange(len(yex)) / sampling_rate
plot(tex, yex);
```

Zoom in:

```python
n = 100000
t = tex[:n]
y1 = yex[:n, 1]
i0 = np.nonzero(np.diff(y1) > 0.3)[0][0] + 1
plot(t, y1, t[i0], y1[i0], 'ro');
```

Simulate some data for testing:

```python
y = np.zeros_like(t)
for i in range(0, 2):
    y[i0 + i*sampling_rate:i0 + (i+0.1)*sampling_rate] = y1.max()
plot(t, y, t, y1); ylim(-1.0, 1.0);
```

High-pass filter

```python
filter_hz = 6.0
bb, ab = signal.butter(1, filter_hz / (sampling_rate / 2.), btype='high')
yf = signal.lfilter(bb, ab, y)
plot(t, yf, t, y1); ylim(-1.1, 1.1); xlim(0.5, 0.8);
```
