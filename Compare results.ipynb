```python
%pylab inline
%load_ext autoreload
%autoreload 2
import pandas as pd
from input import PrerecordedDataSource
from analysis import ClockAnalyser
```

```python
%%capture
def get_data(**options):
    source = PrerecordedDataSource('../../dataq/record_20130331_0002_100s.npz')
    analyser = ClockAnalyser(source)
    return list(analyser.process(**options))

# Read samples, analyze
data = pd.concat([
    pd.DataFrame(get_data(pps_edge='up')).set_index('time'),
    pd.DataFrame(get_data(pps_edge='down')).set_index('time'),
    pd.DataFrame(get_data(pps_edge='up', fit_decay=True)).set_index('time'),
    pd.DataFrame(get_data(pps_edge='down', fit_decay=True)).set_index('time'),
], keys=['up', 'down', 'upfit', 'downfit'], axis=1)

# Correct drift
drift_offset = -2
data.up.drift += drift_offset - 0.1
data.down.drift += drift_offset
data.upfit.drift += drift_offset - 0.1
data.downfit.drift += drift_offset
```

```python
def load_clock_file(date):
    fn = date.strftime('ref_data/clock%Y-%m-%d.txt')
    return pd.read_csv(fn, sep=' ', header=None, names=['time', 'drift', 'amplitude'], index_col=0,
                       date_parser=lambda s: datetime.datetime.fromtimestamp(int(s)), parse_dates=[0])
```
